```python
import pandas as pd

# On charge la base ordonnée
df = pd.read_csv('trafic_2021_ordonne.csv', sep=';')

# Nettoyage final pour l'affichage Jupyter
df['Arrondissement pour Paris'] = df['Arrondissement pour Paris'].fillna('')

# --- VUE 1 : Par Ville ---
view_ville = df.groupby(['Ville', 'Réseau']).agg({
    'Station': 'count',
    'Trafic': 'sum'
}).rename(columns={'Station': 'Nombre de Stations', 'Trafic': 'Trafic Total'}).sort_values('Trafic Total', ascending=False)

# --- VUE 2 : Par Arrondissement (Paris) ---
view_paris = df[df['Ville'] == 'Paris'].groupby('Arrondissement pour Paris').agg({
    'Station': 'count',
    'Trafic': 'sum'
}).sort_values('Trafic', ascending=False)

print("✅ Bases de consultation prêtes !")
```
