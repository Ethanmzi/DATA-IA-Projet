```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# On s'assure que le fichier existe avant de le lire
try:
    df = pd.read_csv('dataset_pret_pour_ml.csv')

    # Visualisation de la premire station
    station_nom = df['station'].unique()[0]
    plt.figure(figsize=(12, 5))

    sns.lineplot(data=df[df['station'] == station_nom], 
                 x='heure', y='affluence', hue='est_weekend')

    plt.title(f"Profil horaire de la station : {station_nom}")
    plt.xlabel("Heure de la journe")
    plt.ylabel("Nombre de voyageurs (estim)")
    plt.legend(title="Week-end", labels=["Semaine", "Week-end"])
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

except FileNotFoundError:
    print("Erreur : Le fichier 'dataset_pret_pour_ml.csv' n'est pas encore cr.")
    print("Vrifie que tu as bien excut la PREMIRE cellule sans erreur.")
```
